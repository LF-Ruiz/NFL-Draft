# Extract

##### Extract the data from Wikipedia and  www.pro-football-reference.com. 
* Note: Since we are scraping data from tables we are using pandas otherwise we were going to need to use BeatifulSoup and Splinter

```python
#url for draft and combine (2016-2020)

combine2020 = 'https://www.pro-football-reference.com/draft/2020-combine.htm#combine'
combine2019 = 'https://www.pro-football-reference.com/draft/2019-combine.htm'
combine2018 = 'https://www.pro-football-reference.com/draft/2018-combine.htm'
combine2017 = 'https://www.pro-football-reference.com/draft/2017-combine.htm'
combine2016 =  'https://www.pro-football-reference.com/draft/2016-combine.htm'
draft2020 = 'https://en.wikipedia.org/wiki/2020_NFL_Draft'
draft2019 = 'https://en.wikipedia.org/wiki/2019_NFL_Draft'
draft2018 = 'https://en.wikipedia.org/wiki/2018_NFL_Draft'
draft2017 = 'https://en.wikipedia.org/wiki/2017_NFL_Draft'
draft2016 = 'https://en.wikipedia.org/wiki/2016_NFL_Draft'
```

```python
# Scrap url's using pandas
combine_2020 = pd.read_html(combine2020)
combine_2019 = pd.read_html(combine2019)
combine_2018 = pd.read_html(combine2018)
combine_2017 = pd.read_html(combine2017)
combine_2016 = pd.read_html(combine2016)
draft_2020 = pd.read_html(draft2020)
draft_2019 = pd.read_html(draft2019)
draft_2018 = pd.read_html(draft2018)
draft_2017 = pd.read_html(draft2017)
draft_2016 = pd.read_html(draft2016)
```

```python
# Check type and lenght of the tables
print(type(combine_2020))
print(len(combine_2020))
print(type(draft_2020))
print(len(draft_2020))
```

```python
# Extract the table needed to get the dataframes
combinedf2020 = combine_2020[0]
combinedf2019 = combine_2019[0]
combinedf2018 = combine_2018[0]
combinedf2017 = combine_2017[0]
combinedf2016 = combine_2016[0]
draftdf2020 = draft_2020[4]
draftdf2019 = draft_2019[4]
draftdf2018 = draft_2018[4]
draftdf2017 = draft_2017[4]
draftdf2016 = draft_2016[4]
```

# Transform

```python
# Add the year 
draftdf2020['year']='2020'
draftdf2019['year']='2019'
draftdf2018['year']='2018'
draftdf2017['year']='2017'
draftdf2016['year']='2016'
combinedf2020['year']='2020'
combinedf2019['year']='2019'
combinedf2018['year']='2018'
combinedf2017['year']='2017'
combinedf2016['year']='2016'
draftdf2016
```

```python
# Append combine dataframes
combine_df = combinedf2020.append(combinedf2019,ignore_index=True,verify_integrity=True)
combine_df = combine_df.append(combinedf2018,ignore_index=True,verify_integrity=True)
combine_df = combine_df.append(combinedf2017,ignore_index=True,verify_integrity=True)
combine_df = combine_df.append(combinedf2016,ignore_index=True,verify_integrity=True)

#combine_df = combine_df.drop(columns = {'Pos'})
combine_df
```

```python
# Append draft dataframes
draft_df = draftdf2020.append(draftdf2019,ignore_index=True,verify_integrity=True)
draft_df = draft_df.append(draftdf2018,ignore_index=True,verify_integrity=True)
draft_df = draft_df.append(draftdf2017,ignore_index=True,verify_integrity=True)
draft_df = draft_df.append(draftdf2016,ignore_index=True,verify_integrity=True)

#draft_df = draft_df.drop(columns = ['Pos'])
draft_df
```

#### Clean dataframes

```python
# Drop columns we are not using and rename for cleaning pourpuses and storage
draft_df = draft_df.drop(columns = ['Notes', 'Unnamed: 0', 'NFL team'])
draft_df = draft_df.rename(columns = {"Pos.": "Pos", "College": "School", "Conf.": "Conf", "Rnd.": "Rnd","Player":"player", 
                                      'Pick #': 'Pick_no'})
combine_df = combine_df.drop(columns = ['Drafted (tm/rnd/yr)', 'College'])
combine_df = combine_df.rename(columns = {'Player': 'player'})


combine_df
```

```python
combine_df["Ht"] = combine_df["Ht"]
def fix(string):
    try:
        feet = int(string.split('-')[0])
    
        inches = int(string.split('-')[1])

    #thats for feet and inches
        return feet * 12 + inches
    except:
        return string
    
combine_df['Ht'] = combine_df['Ht'].apply(fix)
```

```python
combine_df
```

#### Create players dataframes for database

```python
# Create combine player dataframe
#### Note: After a lot of problems with the duplicates, I checked and keeping the position and year in this dataframe 
#### seems the best option. 
player = combine_df['player']
school = combine_df['School']
pos = combine_df['Pos']
year = combine_df['year']
combine_player_df = {"player":player,
                    "pos":pos, 
                    "school":school,
                    "year":year}
combine_player_df = pd.DataFrame(combine_player_df)
combine_player_df
```

```python
# Create draft player dataframe
player = draft_df['player']
school = draft_df['School']
pos = draft_df['Pos']
year = draft_df['year']
draft_player_df = {"player":player,
                  "pos":pos,
                  "school":school,
                  "year":year}
draft_player_df = pd.DataFrame(draft_player_df)
draft_player_df
```

```python
# Append both players dataframes
#### Note: Set ignore index so the index show the total players, the others two arguments could or could be not set.
#### We can also use concat, but concat doesn't have ignore index argument. 
players_df = draft_player_df.append(combine_player_df, ignore_index=True, verify_integrity=True, sort=True)
players_df
```

```python
# Drop the duplicates in players_df.
##### Note: set subset='player' so is going to drop the duplicates in the players column, that way we can keep position
##### for analysis pourposes. After this, we should have 2008 unique players

players_df = players_df.drop_duplicates(subset='player', keep='first', ignore_index=True)
players_df
```

```python
# Check the dataframe to see if there are still duplicates
print(players_df.loc[players_df['player']== 'Anthony Zettel'])
print(players_df.nunique(axis = 0, dropna=False))
```

```python
# Append data using pandas. name stands for table name, change it if neccesary.
players_df.to_sql(name='player', con=engine, if_exists='append', index=False)
```

```python
#Confirm the data has been appended. Check table name. Create a dataframe to get player_id to the other dataframes
player_id_df=pd.read_sql_query('select * from player', con=engine)
player_id_df
```

```python
# merge with combine dataframe to get the serial id.
combine_df_final = pd.merge(combine_df, player_id_df, how='inner', on='player')
combine_df_final
```

```python
# Clean the combine df final by dropping columns we don't need and renaming id

combine_df_final = combine_df_final.drop(columns = ['school', 'School','player', 'year_y', 'pos', 'Pos', 'year_x'])
combine_df_final = combine_df_final.rename(columns = {'id':'player_id'})

# Convert types to fit in the database
#### we are using to numeric for floats and convert_dtypes method for string
#### Height needs to be string because is in 6 feet 1 inch format. others are float
combine_df_final['Ht'] = combine_df_final['Ht'].convert_dtypes(infer_objects=True, convert_string=True)
combine_df_final['Wt'] = pd.to_numeric(combine_df_final['Wt'], errors='coerce')
combine_df_final['40yd'] = pd.to_numeric(combine_df_final['40yd'], errors='coerce')
combine_df_final['Bench'] = pd.to_numeric(combine_df_final['Bench'], errors='coerce')
combine_df_final['Vertical'] = pd.to_numeric(combine_df_final['Vertical'], errors='coerce')
combine_df_final['Broad Jump'] = pd.to_numeric(combine_df_final['Broad Jump'], errors='coerce')
combine_df_final['3Cone'] = pd.to_numeric(combine_df_final['3Cone'], errors='coerce')
combine_df_final['Shuttle'] = pd.to_numeric(combine_df_final['Shuttle'], errors='coerce')
# Get the info of the dataframe
combine_df_final.info()
```

```python
# merge with draft dataframe to get the serial id.
draft_df_final = pd.merge(draft_df, player_id_df, how='inner', on='player')
draft_df_final
```

```python
# Clean the draft df final by dropping columns we don't need and renaming id

draft_df_final = draft_df_final.drop(columns = ['school','School','player', 'year_y', 'pos', 'Pos', 'year_x'])
draft_df_final = draft_df_final.rename(columns = {'id':'player_id'})

# # Convert types to fit in the database
### Rnd coulb be either a string or a float because we are not supossed to do any math operation with it. 
draft_df_final['Rnd'] = pd.to_numeric(draft_df_final['Rnd'], errors='coerce')
draft_df_final['Pick_no'] = draft_df_final['Pick_no'].convert_dtypes(infer_objects=True, convert_integer=True)
draft_df_final['Conf'] = draft_df_final['Conf'].convert_dtypes(infer_objects=True, convert_string=True)


draft_df_final.info()
```

```python
# Load data using pandas. name stands for table name, change it if neccesary.
combine_df_final.to_sql(name='combine', con=engine, if_exists='append', index=False)
```

```python
# Load data using pandas. name stands for table name, change it if neccesary.
draft_df_final.to_sql(name='draft', con=engine, if_exists='append', index=False)
```
